#  Background Information



In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participantâ€™s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

$\textbf{Answer}$
1. The independent variable in the experiment is whether the words are congruent or incongruent and the dependent variable is the amount of time taken to name the color of the words.

2. Let $\mu_{C}$ be the  population mean of the reaction time for naming congruent words;

   Let $\mu_{IC}$ be the  population mean of the reaction time for naming incongruent words;
   
   The appropriate set of  hypotheses for this experiment can be given by:
   
Null Hypothesis (H$_{O}$):

The average population reaction time to name the colors is the same for $\textit{congruent}$ and $\textit{incongruent}$ words.

Alternate Hypothesis (H$_{A}$):

The average population reaction time to name the colors is different for $\textit{congruent}$ and $\textit{incongruent}$ words.

$\textbf{Mathematically;}$
   
   $H_{O}$ : $\mu_{C}$= $\mu_{IC}$
   
   $H_{A}$ : $\mu_{C}$ $\neq$ $\mu_{IC}$
   
 Since the two tests are performed on the same subject (participants) but with two different conditions $\textit{i.e}$,$\textbf{Congruent}$ and $\textbf{Incongruent}$ words the samples are paired. More over the sample size is 25 which is small. For the two  mentioned reasons the appropriate  statistical test for this experiment is dependent two-tailed paired T-test.

# 4. Statistical Test
The appropriate set of  hypotheses for this experiment are:
   
   $H_{O}$ : $\mu_{C}$= $\mu_{IC}$
   
   $H_{A}$ : $\mu_{C}$ $\neq$ $\mu_{IC}$
   
Let,
     $X_{C}$ = The sample mean of the reaction time for congruent words, and 
     $X_{IC}$ = The sample mean of the reaction time for incongruent words;
     
     
  Then  the point estimate of the the population mean difference $\mu_{C}$$-$$\mu_{IC}$ = $X_{C}$$-$$X_{IC}$ =7.96479166667.

  Our T-statistics test then given by:  
  
  t = ($X_{C}$$-$$X_{IC}$)/($S_{D}$/$\sqrt n$)
  
 Where n =25, is our sample size, and $S_{D}$ is the standard error of the difference of the sample calculated below.
  
 Lets first create another column (the difference) in our data frame to calculate the standard error $S_{D}$.
 

```python
data["difference"]=data["Incongruent"]-data["Congruent"]
data.head()
SD=data["difference"].std()
D= data["difference"].mean()
print "The standard error of the difference =", SD
t=(data["Incongruent"].mean()-data["Congruent"].mean())/(SD/np.sqrt(24))
print "The t_value = %4.2f" % t
#print "The mean of the difference =",D
#SD=np.sqrt(np.sum((data["difference"]-data["difference"].mean())**2)/23)
#t=(data["Incongruent"].mean()-data["Congruent"].mean())/(SD/np.sqrt(24))
```

With 95 % confidence level, the critical t value ($t_{Critical}$$=$$\pm$ 2.064) and our two-tailed coupled  T-test value is 8.02 which lies in the rejection region. Hence, we reject the null hypothesis the population means of the reaction times  to name congruent and incongruent words are the same ($H_{O}$). This shows there is significant difference in the population mean of the reaction times to name the colors of words when the words are congruent and incongruent. The test result is consistent with our previous analysis and expected since our results from the calculated central tendencies of the samples show significant differences.  

This result shows that  partcipants react fast  to name the colors if the words  are congruent. The reaction time might also be affected by the reading ability of participants. Participants with reading problems might react slowly compared to others in congruent words.
